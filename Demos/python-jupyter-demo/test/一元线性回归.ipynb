```python
import pandas as pd
data = pd.read_csv("../data/武汉市房地产数据.csv", index_col=0)
data = data.T
data = data.fillna(method='bfill')
data
```

```python
data_chg = (data-data.shift(1))/data.shift(1)
data_chg = data_chg.iloc[1:]
data_chg
```

```python
# data1 = (data-data.shift(1))/data.shift(1)
# data1
```

```python
import scipy.stats as stats
x = data_chg["个人住房贷款额（亿元）"]
y = data_chg["房价（元／平）"]
r = stats.pearsonr(x, y)[0]
r
```

```python
result = stats.linregress(x, y)
result
```

```python
print("斜率: %.2f" %result.slope)
print("截距: %.2f" %result.intercept)
print("拟合优度R-squared: %.2f" % result.rvalue**2)
print("斜率估计量的标准误差: %.2f" % result.stderr)
print("截距估计量的标准误差: %.2f" % result.intercept_stderr)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x, y, label='original data', linewidths=4)
plt.plot(x, result.intercept + result.slope*x, label='fitted line', linewidth=2, color="red")
plt.legend()
```
