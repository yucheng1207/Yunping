## 3. 使用模版采集数据

八爪鱼官方提供的、做好的采集模板，目前已有200+采集模板，涵盖主流网站的采集场景。

使用模板采集数据时，只需输入几个参数（网址、关键词、页数等），就能在几分钟内快速获取到目标网站数据。（类似PPT模板，只需修改关键信息就能直接使用，无需自己从头配置。）

### 浏览热门模版

![img](图片/2.gif)

### 通过搜索找到模版

在客户端首页【输入框】中，输入目标网站名称，八爪鱼自动寻找相关的采集模板。将鼠标移到需要的模板上并单击，进入模板详情页面。

注意，请确保输入的网站名称正确，否则可能无法查找到相关模板。

我们在搜索框中输入：京东商品详情采集，然后，点击选择。

![Screen Shot 2021-09-14 at 5.46.57 PM](图片/57PM.png)



### 使用模版

第一步，进入【模板详情页】后，仔细阅读【模板介绍】、【采集字段预览】、【采集参数预览】、【示例数据】，确认此模板采的数据符合需求。

![Screen Shot 2021-09-14 at 5.47.20 PM](图片/20PM.png)



第二步，确定模板符合需求以后，点击【立即使用】，自行【配置参数】。常见的参数有关键词、翻页次数、URL等。请认真查看【模板介绍】中的使用方法说明和参数说明，输入格式正确的参数，否则将影响模板的使用。

在输入商品详情页面网址这一栏输入： https://item.jd.com/100016034406.html



第三步，然后点击【保存并启动】，选择启动【本地采集】。八爪鱼自动启动1个采集任务并采集数据。

![Screen Shot 2021-09-14 at 5.48.59 PM](图片/59PM.png)

第四步，数据采集完成以后，可以需要的格式导出。采集过程中可以随时停止，并保存已采集的数据。

![Screen Shot 2021-09-14 at 5.52.30 PM](图片/1613916.png)

## 3.1.4 自定义采集任务

示例网址：https://mall.ebaiyin.com/category_3.shtml

第一步，在首页输入框中，输入目标网址，点击【开始采集】。八爪鱼自动打开网页并开始智能识别。

![img](图片/1-20210914132310485.gif)

给它一点时间，等待智能识别完成。

![img](图片/2-20210914132356470.gif)

第二步，智能识别成功，一个网页可能有多组数据，八爪鱼会将所有数据识别出来，然后智能推荐最常用的那组。如果推荐的不是想要的，可自行【切换识别结果】，这里我们不点击，看第三步。

![img](图片/7.png)

第三步：同时，可自动识别出网页的滚动和翻页。此示例网址，无需滚动，只需翻页，故只识别并勾选【翻页采集】。

![img](图片/8.png)

第四步：自动识别完成后，点击【生成采集设置】，可自动生成相应的采集流程，方便用户编辑修改。

![Screen Shot 2021-09-14 at 6.08.33 PM](图片/33PM.png)

第五步，然后，点击左上角的【采集】，选择【启动本地采集】，八爪鱼就会开始全自动采集数据。

第六步：采集完成后，以所需的方式导出数据即可。

![img](图片/11.png)
