```python
import pandas as pd
pd.set_option('mode.chained_assignment', None)
data = pd.read_csv('data/world-happiness-report-china.csv')
# data
```

```python
data.duplicated() # 检查哪些行有重复值
```

```python
data = data.drop_duplicates() # 对数据集中重复的行进行去重操作
```

```python
data.isnull() # 检查是否有缺失值
```

```python
# data.isnull().any() # 按照列的方向，观察是否有缺失值
```

```python
data = data.fillna(0) # 使用0来替代缺失值
```

```python
col = data["Life Ladder"] # 将Life Ladder列中所有大于5的值都替换成5
data["Life Ladder"]
col[col>5.0]=5.0
col.describe()
```
